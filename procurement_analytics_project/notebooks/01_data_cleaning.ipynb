```python
df_clean = df.copy()
```

```python
df_clean.columns = (
    df_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
    .str.replace("/", "_")
)
df_clean.columns
```

```python
text_cols = df_clean.select_dtypes(include="object").columns

for col in text_cols: 
        df_clean[col] = df_clean[col].astype(str).str.strip()
```

```python
duplicates_before = df_clean.duplicated().sum()
print("Duplikate vorher:", duplicates_before)

df_clean = df_clean.drop_duplicates()

duplicates_after = df_clean.duplicated().sum()
print("Dublikate nachher:", duplicates_after)
```

```python
df_clean["order_date"] = pd.to_datetime(df_clean["order_date"], errors="coerce")
df_clean["delivery_date"] = pd.to_datetime(df_clean["delivery_date"], errors="coerce")

numeric_cols = ["quantity", "unit_price", "negotiated_price", "defective_units"]

for col in numeric_cols:
    df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")
```

```python
df_clean["order_year"] = df_clean["order_date"].dt.year
df_clean["order_month"] = df_clean["order_date"].dt.month
df_clean["year_month"] = df_clean["order_date"].dt.to_period("M").astype(str)

df_clean["lead_time_days"] = (df_clean["delivery_date"] - df_clean["order_date"]).dt.days

df_clean["original_spend"] = df_clean["quantity"] * df_clean["unit_price"]
df_clean["negotiated_spend"] = df_clean["quantity"] * df_clean["negotiated_price"]

df_clean["savings"] = df_clean["original_spend"] - df_clean["negotiated_spend"]

df_clean["savings_rate"] = df_clean["savings"] / df_clean["original_spend"]

df_clean["defect_rate"] = df_clean["defective_units"] / df_clean["quantity"]
  
```

```python
df_clean["is_delivered"] = (df_clean["order_status"] == "Delivered").astype(int)
df_clean["is_partially_delivered"] = (df_clean["order_status"] == "Partially Delivered").astype(int)
df_clean["is_pending"] = (df_clean["order_status"] == "Pending").astype(int)
df_clean["is_cancelled"] = (df_clean["order_status"] == "Cancelled").astype(int)

df_clean["is_completed_delivery"] = (
    df_clean["order_status"].isin(["Delivered", "Partially Delivered"])
).astype(int)

df_clean["is_compliant"] = (df_clean["compliance"] == "Yes").astype(int)
```
